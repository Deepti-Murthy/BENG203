```python
import pandas as pd

cancer_df = pd.read_csv("C:/Users/noapo/Downloads/pnas_tpm_96_nodup.txt", sep="\t", header=None)
cancer_df.set_index(cancer_df.columns[0], inplace=True)
cancer_df.columns = [f'C{i+1}' for i in range(cancer_df.shape[1])]
cancer_df.head()
```

```python
oncogenes_df = pd.read_csv("C:/Users/noapo/Downloads/cancerGeneList.tsv", sep="\t")
oncogenes_df.head()
```

```python
from pybiomart import Dataset

dataset = Dataset(name='hsapiens_gene_ensembl', host='http://www.ensembl.org')
mapping_df = dataset.query(attributes=['ensembl_gene_id', 'hgnc_symbol'])
mapping_df.columns = ['Gene stable ID', 'HGNC symbol']
mapping_df = mapping_df[mapping_df['HGNC symbol'] != '']

ensg_to_symbol = dict(zip(mapping_df['Gene stable ID'], mapping_df['HGNC symbol']))

cancer_df['gene_symbol'] = cancer_df.index.map(ensg_to_symbol)
cancer_df = cancer_df[cancer_df['gene_symbol'].notnull()]
cancer_df.set_index('gene_symbol', inplace=True)
cancer_df.head()
```

```python
oncogene_genes = oncogenes_df['Hugo Symbol'].tolist()
filtered_df = cancer_df.loc[cancer_df.index.intersection(oncogene_genes)]
filtered_df.head()
```

```python
metadata_df = pd.read_csv("C:/Users/noapo/Downloads/pnas_patient_info.csv")

column_names = [f"C{i+1}" for i in range(len(metadata_df))]
recur_statuses = metadata_df['recurStatus'].tolist()
new_column_names = [f"{name}_{status}" for name, status in zip(column_names, recur_statuses)]

filtered_df.columns = new_column_names
filtered_df.head()
```

```python
import pandas as pd

filtered_columns = filtered_df.columns.tolist()

metadata_df['base_id'] = [f"C{i+1}" for i in range(len(metadata_df))]
metadata_df['full_id'] = metadata_df['base_id'] + '_' + metadata_df['recurStatus']
metadata_df.set_index('full_id', inplace=True)

aligned_metadata = metadata_df.loc[filtered_df.columns]
aligned_metadata['recurStatus'] = aligned_metadata['recurStatus'].replace({'R': 1, 'N': 0}).astype(int)

metadata_fields = ['fu', 'cancerstage_cat', 'er_cat', 'pr_cat', 'her2_cat', 'recurStatus']
metadata_rows = aligned_metadata[metadata_fields].T

combined_with_metadata = pd.concat([filtered_df, metadata_rows])
combined_with_metadata.tail(10)
```
